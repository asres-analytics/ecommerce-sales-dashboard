##  Step 5: Remove Cancelled Transactions

In this step, I remove cancelled orders from the dataset.

###  Why this is important:
- Invoice numbers starting with "C" represent cancelled transactions
- These are not actual sales and should not be included in revenue calculations
- Keeping them would lead to incorrect analysis and misleading insights

 Next: I will identify and remove all cancelled transactions.


```python
# Find cancelled transactions
cancelled = df[df['InvoiceNo'].astype(str).str.startswith('C')]

cancelled.shape
```

```python
# Remove cancelled transactions
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# Check again
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

All cancelled transactions have been successfully removed.

The dataset now contains only valid sales transactions, ensuring accurate analysis going forward.
